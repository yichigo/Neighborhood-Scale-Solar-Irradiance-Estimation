```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter

```

```python
dir_data = '../data/LISIRD/'
fn = 'nrl2_tsi_P1D.csv'
dir_out = '../figures/season/'

df = pd.read_csv(dir_data + fn)
df = df.rename(columns={df.columns[0]: "date"})

start = datetime.datetime(1882 ,1, 1)
end = datetime.datetime(2019 , 12, 31)
df['date'] = pd.date_range(start, end, freq='1d')
df = df.set_index('date')
df
```

```python
plt.rcParams["figure.figsize"] = (20, 8) # (w, h)
plt.rcParams.update({'font.size': 20})
fig, ax = plt.subplots()

plt.scatter(df.index, df.iloc[:,0], s=0.1, c = 'black',label = '')
plt.title('Naval Research Laboratoryâ€™s Solar Variability Model V2 (NRLTSI2) \nDaily Average of Total Solar Irradiance (TSI): 1882-2019')
plt.xticks(rotation=45)
# plt.xlabel('date') 
plt.ylabel('TSI $(W/m^2)$')
plt.gca().relim()

date_form = DateFormatter("%Y")
ax.xaxis.set_major_formatter(date_form)
ax.xaxis.set_major_locator(mdates.YearLocator(10))
plt.grid()
plt.tight_layout()
plt.savefig(dir_out+'TSI_daily_1882_2019.png')
```

```python
dir_data = '../data/LISIRD/'
fn = 'nrl2_tsi_P1Y.csv'
dir_out = '../figures/season/'

df = pd.read_csv(dir_data + fn)
df = df.rename(columns={df.columns[0]: "date"})

df['date'] = list(range(1610,2019+1))
df = df.set_index('date')
df
```

```python
plt.rcParams["figure.figsize"] = (20, 8) # (w, h)
plt.rcParams.update({'font.size': 20})
fig, ax = plt.subplots()

plt.plot(df.index, df.iloc[:,0], color = 'k', marker = 'o', linestyle='dashed', linewidth=1, markersize=4)
plt.title('Naval Research Laboratoryâ€™s Solar Variability Model V2 (NRLTSI2) \nYearly Average of Total Solar Irradiance (TSI): 1610-2019')
plt.xticks(rotation=45)
# plt.xlabel('date') 
plt.ylabel('TSI $(W/m^2)$')
plt.gca().relim()

plt.grid()
plt.tight_layout()
plt.savefig(dir_out+'TSI_yearly_1610_2019.png')
```

```python
dir_data = '../data/LISIRD/'
fn = 'tsis_tsi_24hr.csv'
dir_out = '../figures/season/'

df = pd.read_csv(dir_data + fn)
df = df.rename(columns={df.columns[0]: "date"})

start = datetime.datetime(2018 ,1, 11)
end = datetime.datetime(2020 , 12, 31)
df['date'] = pd.date_range(start, end, freq='1d')
df = df.set_index('date')


iwant = (df.iloc[:,0] > 0)
df = df[iwant]
df
```

```python
plt.rcParams["figure.figsize"] = (20, 8) # (w, h)
plt.rcParams.update({'font.size': 20})
fig, ax = plt.subplots()

plt.plot(df.index, df.iloc[:,0], color = 'k', marker = 'o', linestyle='dashed', linewidth=1, markersize=4)
plt.title('LASP TSIS Spectral Irradiance Monitor (SIM) \nDaily Average of Total Solar Irradiance (TSI): 2018-2020')
plt.xticks(rotation=45)
# plt.xlabel('date') 
plt.ylabel('TSI $(W/m^2)$')
plt.ylim([1361,1362])
date_form = DateFormatter("%Y-%m")
ax.xaxis.set_major_formatter(date_form)
ax.xaxis.set_major_locator(mdates.MonthLocator([1,4,7,10]))
plt.grid()
plt.tight_layout()
plt.savefig(dir_out+'TSI_daily_2018_2020.png')
```
